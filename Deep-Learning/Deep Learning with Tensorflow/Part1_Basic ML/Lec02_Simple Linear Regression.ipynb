# Build Hypothesis and Cost

```python
import tensorflow as tf

x_data = [1,2,3,4,5]
y_data = [1,2,3,4,5]

# 임의의 초기 계수
W = tf.Variable(2.9)
b = tf.Variable(0.5)

# Hypothesis(가설)
hypothesis = W * x_data + b

# Cost function(비용함수)
cost = tf.reduce_mean(tf.square(hypothesis - y_data))
```

# Gradient Descent Algorithm
- Algorithm of Minimizing Cost

```python
# learning_rate initialize
learning_rate = 0.01

# Gradient descent
with tf.GradientTape() as tape:
    # 변수들(W, b)의 변화에 대한 정보를 tape에 기록함
    hypothesis = W * x_data + b
    cost = tf.reduce_mean(tf.square(hypothesis - y_data))

# cost함수에 대해서 변수들[W, b]에 대한 개별 기울기(경사도,미분값)를 튜플로 반환 
W_grad, b_grad = tape.gradient(cost, [W, b])

# W와 b값을 업데이트
# A.assign_sub(B): A = A - B
# learning_rate: 기울기 값을 얼마만큼 반영할 것인지 결정하는 계수
W.assign_sub(learning_rate * W_grad)
b.assign_sub(learning_rate * b_grad)
```

# Parameter(W, b) Upadte
- Epoch
- Training


```python
import tensorflow as tf
# 즉시실행
tf.compat.v1.enable_eager_execution() # tf.enable_eager_execution()

# Data
x_data = [1,2,3,4,5]
y_data = [1,2,3,4,5]

# W, b initialize(random)
W = tf.Variable(2.9)
b = tf.Variable(0.5)

learning_rate = 0.01

print("epoch | W | b | cost")
for epoch in range(101):
    # Gradient descent
    with tf.GradientTape() as tape:
        hypothesis = W * x_data + b
        cost = tf.reduce_mean(tf.square(hypothesis - y_data))
    
    W_grad, b_grad = tape.gradient(cost, [W, b])

    W.assign_sub(learning_rate * W_grad)
    b.assign_sub(learning_rate * b_grad)

    if epoch % 10 == 0:
        print("{:5}|{:10.4}|{:10.4}|{:10.6f}".format(epoch, W.numpy(), b.numpy(), cost))
```
